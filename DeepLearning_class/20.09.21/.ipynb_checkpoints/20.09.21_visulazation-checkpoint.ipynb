```python
data = pd.DataFrame({
    'food':['bacon','pork','bacon','beef','Bacon','ham'],
    'ounces':[4,3,12,6,8,3]
})
data
```

```python
# mapping rule
meat_to_animal={
    'bacon':'pig',
    'pork':'pig',
    'beef':'cow',
    'ham':'pig'
}
```

```python
# data['food'] 소문자 변환
# 시리즈는 안돼!
# 중간에 str을 써서.
lowercased = data['food'].str.lower()
```

```python
# meat_to_animal를 참조하여 lowercased에 저장된 값에 대한 대응되는 값 출력 -> map 함수!!
# 시리즈.map(딕셔너리)  ->  시리즈 타입으로 저장된 데이터가 딕셔너리의 키가 되며, 키에 해당되는 값 출력
data['animal'] = lowercased.map(meat_to_animal)
```

```python
data
```

```python
# lambda 함수 활용
data = pd.DataFrame({
    'food':['bacon','pork','bacon','beef','Bacon','ham'],
    'ounces':[4,3,12,6,8,3]
})
data['food'].map(lambda x: meat_to_animal[x.lower()])
```

```python
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(nrows = 5)
fig.set_size_inches(18,25)
sns.pointplot(data=train, x='hour',y = 'count', ax = ax1)
sns.pointplot(data=train, x='hour',y = 'count', hue = 'workingday',ax = ax2)
sns.pointplot(data=train, x='hour',y = 'count', hue = 'dayofweek',ax = ax3)
sns.pointplot(data=train, x='hour',y = 'count', hue = 'weather',ax = ax4)
sns.pointplot(data=train, x='hour',y = 'count', hue = 'season',ax = ax5)
```

```python
train.columns
# corrlation matrix
corrMatt = train[['temp','season','casual','registered','humidity','windspeed','count']]
corrMatt = corrMatt.corr()

mask = np.array(corrMatt)
mask[np.tril_indices_from(mask)] = False
```

```python
fig.ax = plt.subplots()
fig.set_size_inches(20,10)
sns.heatmap(corrMatt,square = True, annot = True, vmax = 0.8, mask = mask)
```

```python
fig,(ax1,ax2,ax3)=plt.subplots(ncols = 3) # nrows = 3
fig.set_size_inches(12,5)
sns.regplot(x='windspeed',y='count',data = train, ax = ax1)
sns.regplot(x='humidity',y='count',data = train, ax = ax2)
sns.regplot(x='temp',y='count',data = train, ax = ax3)
```

```python
def cym(datetime):
    return "{}-{}".format(datetime.year, datetime.month)
train['ym'] = train['datetime'].apply(cym)
train[['datetime','ym']]
fig, (ax1, ax2) = plt.subplots(ncols = 2)
fig.set_size_inches(18,4)
sns.barplot(data=train, x = 'year', y = 'count', ax = ax1)
sns.barplot(data=train, x='month',y = 'count',ax = ax2)
fig, ax3 = plt.subplots()
fig.set_size_inches(18,4)
sns.barplot(data=train, x = 'ym', y = 'count', ax = ax3)
```
