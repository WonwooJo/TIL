```python
# 연관규칙(중요!)
'''
연관규칙: 머신러닝 관점에서 비지도학습.
머신러닝: 기계를 학습시킴. 마치 인간처럼 어떤 문제가 주어지면 문제에 대한 답을 스스로 고민하고 판단하도록 하는 능력 부여
- 지도학습: Xdata, Ydata 모두 존재.
- 비지도학습: 연관규칙
'''
```

```python
'''
상품 추천 분석 기법?
1) 연관규칙분석(장바구니 분석): 기존 고객의 대규모 거래 데이터 존재
                -> 함께 구매가 발생하는 규칙 발견
                => 고객이 특정 상품 구매 시 이와 연관성이 높은 다른 상품을 추천하자!

2) 협업 필터링: 고객의 상품 구매 이력 데이터를 수치화
                -> 추천 대상 고객 A와 다른 고객 B에 대한 상관계수.
                => 서로 높은 상관관계에 해당되는 고객 B가 구매한 상품 중 고객 A가 구매한 적 없는 상품을
                   고객 A에게 추천!

3) 컨텐츠 기반 추천 시스템: 고객이 과거에 구매했던 상품들의 속성과 유사한 다른 상품 중 
                           미 구매상품 추천(상품 속성기반 추천)                        
'''
```

```python
'''
연관규칙
규칙?(rules): if 조건식 then 결과 (if A -> B)
연관규칙?: 특정 사건 발생 -> 함께 발생하는 다른 사건 간의 규칙.
          - 특정 항목 집합이 발생했을 떄 또 다른 항목 집합이 발생하는 규칙
항목집합(item set): 전체 항목 집합 중에서 만들어질 수 있는 모든 각각의 부분 집합
x = {a,b,c}
x의 부분집합: 2^3개
의미있는 연관규칙 발견 -> 기준: support(지지도), confidence(신뢰도), lift(향상도)
                            => 특히 향상도가 중요!

'''
```

```python
'''
X와 Y는 공통 원소가 없는 항목 집합이다.
X -> Y: if X then Y 라는 연관규칙 (X를 사는 사람은 Y도 사더라)\
N: 전체 거래 건 수
n(X): 항목 집합 X의 거래 건 수
n(Y): 항목 집합 Y의 거래 건 수

1) 지지도?
두 항목 집합 X와 Y의 지지도?
전체 거래 건 수 중에서 항목 집합 X와 Y를 모두 포함하는 거래건 수의 비율
*표기 예) s(X->Y): X를 구매하면 Y도 구매한다라는 규칙에 대한 지지도
                   X와 Y를 모두 포함하는 거래건수 / 전체 거래건수
                
2) 신뢰도?
항목 집합 X를 포함하는 거래(n(X)) 중에서 항목 집합 Y를 포함하고 있는 거래의 비율(조건부 확률)
*표기 예) c(X->Y) = X와 Y를 모두 포함하는 거래건수 / 항목 집합 X의 거래건수

3) 향상도?
항목 집합 X가 주어지지 않았을때의 항목집합 Y의 확률(분모) 대비 
항목 집합 X가 주어졌을 때의 항목집합 Y의 확률 증가 비율
=> 향상도가 1보다 크거나(+관계) 작다면(-관계) 우연적 기회(random chance)보다 우수함을 의미함.
(X와 Y가 서로 독립이면 Lift = 1)
*표기 예) Lift(X->Y) = c(X->Y) / s(Y)
        Lift(X->Y): 연관규칙 신뢰도 / 지지도  = c(X->Y) / s(Y)
s(Y): 전체 거래중에서 Y거래 건수의 비율



'''
```
