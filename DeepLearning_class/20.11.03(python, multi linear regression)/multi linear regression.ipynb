### 신경망 기반 임의의 중간/기말/가산점이 입력됐을때 최종 점수 출력

```python
import numpy as np
```

```python
from keras.models import Sequential # 모델 생성 클래스 # keras.io
from keras.layers import Dense # Layer 추가 클래스
from keras import optimizers # 최적화 함수 객체
```

```python
# 중간, 기말, 가산점
x=np.array([[70,85,11],[71,89,18],[50,80,20],[99,20,10],[50,10,10]]) 

# 최종점수 
y=np.array([73,82,72,57,31])
```

```python
model = Sequential()
```

```python
model.add(Dense(1,input_dim = 3,activation='linear'))
```

```python
sgd = optimizers.SGD(lr=0.00001) # learning rate를 매우 작게하면!! 
# w:= lr*미분값인데! 발산을 최소화하고자 함
```

```python
model.compile(optimizer=sgd, loss='mse',metrics=['mse'])
```

```python
model.fit(x,y,batch_size=1,epochs=1000)
```

```python
model.predict(x)
```

```python
xTest = np.array([[40,50,20],[30,100,10]])
model.predict(xTest)
```

```python
from keras.datasets import imdb
```

```python
# top 10,000개 
(trainData, trainLabel), (testData, testLabel) = imdb.load_data(num_words=10000)
```

```python
trainData.shape # 리뷰가 25,0000개, 단어는 10,000개
trainLabel.shape
testData.shape
testLabel.shape
```

```python
max([max(seq) for seq in trainData])
```

```python
# 데이터셋 단어와 인덱스
word_index = imdb.get_word_index()
```

```python
rev_word_index = dict((value, key) for (key, value) in word_index.items())
```

```python
rev_word_index[1]
```

```python
# imdb dataset에는 0:padding, 1:문서시작, 2:사전에 없음(OOV) 약속 인덱스
# 문서 시작(1)은 왜 넣는걸까..?
# LSTM같은 알고리즘 돌릴때 걔가 안대...
decodedReview = ' '.join([rev_word_index.get(i-3,'?') for i in trainData[0]])
decodedReview
```

```python
def vs(seq,dimension=10000):
    res = np.zeros((len(seq),dimension)) # 크기가 len(seq), 10000인 모든 원소가 0인 행렬
    # zeros 25000 * 10000 행렬
    for i,sequence in enumerate(seq):
        res[i, sequence] = 1
                   
    return res

xTrain = vs(trainData)
xTest = vs(testData)
```

```python
xTrain.shape
```

```python
xTrain[0]
yTrain = trainLabel.astype('float32') # 신경망 기본 타입은 float
yTest = testLabel.astype('float32')
```

```python
model = Sequential()
```

```python
model.add(Dense(64,input_dim = 10000, activation='relu')) # 출력을 32개로 하겠다.
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='sigmoid')) # 다중분류면 softmax를 쓰면 돼
```

```python
# 모델의 전체적인 구조
model.summary() # 320000 + 32(bias)
```

```python
model.compile(optimizer='rmsprop',
             loss = 'binary_crossentropy',
             metrics=['acc']) # optimizer: adam을 제일 많이 씀
```

```python
xTrain.shape
```

```python
# 검증용으로 데이터 나누기
xVal = xTrain[:10000] # validation
```

```python
partialXTrain = xTrain[10000:] # train
```

```python
yVal = yTrain[:10000]
```

```python
partialYTrain = yTrain[10000:]
```

```python
history=model.fit(partialXTrain, partialYTrain, epochs=20, batch_size = 512,
         validation_data=(xVal,yVal)) # batch_size는 보통 2^n의 형태로 줌(속도가 더 빠르대)
```

```python
# acc : 모의고사(train)
# val.acc : 작년 수능(validation)
# test: 올해 수능(test)
```

```python
# 오버피팅 확인

historyDict = history.history
```

```python
historyDict.keys()
```

```python
loss = historyDict['loss']
acc = historyDict['acc']
val_loss = historyDict['val_loss']
val_acc = historyDict['val_acc']
```

```python
import matplotlib.pyplot as plt
```

```python
epochs = range(1,len(acc)+1)
```

```python
plt.plot(epochs,loss,'bo',label='training cost') # 파란색 점
plt.plot(epochs,val_loss,'b',label='validation cost') # 파란색 점
plt.legend()
plt.show()
```

```python
plt.clf()
```

```python
plt.plot(epochs,acc,'bo',label='training acc') # 파란색 점
plt.plot(epochs,val_acc,'b',label='validation acc') # 파란색 점
plt.legend()
plt.show()
```

```python
# 출력 노드 갯수 변경 (32, 64 등)
# 계층 추가
# 활성화함수 변경 → tanh(-1 ~ +1)
model = Sequential()
model.add(Dense(16,input_dim = 10000, activation='relu')) # 출력을 32개로 하겠다.
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid')) # 다중분류면 softmax를 쓰면 돼
```

```python
model.compile(optimizer='rmsprop',
             loss = 'binary_crossentropy',
             metrics=['acc']) # optimizer: adam을 제일 많이 씀
```

```python
history=model.fit(partialXTrain, partialYTrain, epochs=4, batch_size = 512) # batch_size는 보통 2^n의 형태로 줌(속도가 더 빠르대)
```

```python
model.predict(xTest) # 0.51999164: 긍정과 부정의 단어가 뒤섞여있나보다
```
