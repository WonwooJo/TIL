### 이항변수화 연습

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np
```

```python
# 학생 ID, 성별, 연령대, 성적\
'''
   sn_1, m, 30대(1), A
    .. , f, 20대(0), B
   sn_5, m, 40대(2), F
''' # 성별(0, 1), 연령(0, 1, 2), 성적(0, 1, 2, 3, 4)
data = np.array([[0,0,0],
         [0,1,1],
         [0,2,2],  # 자릿수: 01 012 01234 
         [1,0,3],  # 데이터: 01 100 00010 (맞는 부분만 1로 변환, 나머지 0)
         [1,1,4]]) # 데이터: 01 010 00001
data
```

```python
ohe = OneHotEncoder()
```

```python
ohe.fit(data) # ohe를 data에 fitting 시킴
```

```python
ohe.transform(data)
# sparse matrix: 귀소행렬 (대부분 값이 0)
# <-> ??
```

```python
ohe.transform(data).toarray()
```

```python
# Binarizer: 연속형 데이터 -> 이항변수화 or digitize(), where()
# OneHotEncoder: 범주형 데이터 -> 이항변수화
# 이산형화: 연속형 데이터 -> 2개 이상의 범주를 갖는 변수로 변환하는 것
```

```python
'''
ex) 점수(95) -> 90이상: A, 80이상: B, ... 60미만: F
'''
```

```python
np.random.seed(11)
```

```python
df = pd.DataFrame({
    'c1': np.random.randn(20),
    'c2': ['a']*10+['b']*10   # ['a',...'a','b',...,'b'] a: 10개, b: 10개
})
```

```python
df
```

```python
df.c1.max()
```

```python
df.c1.min()
```

```python
# 0~10까지 5구간으로 균등하게 나눔
np.linspace(0,10,5)
```

```python
# 최소~쵀대값을 9개의 구간으로 나눔. 
bins = np.linspace(df.c1.min(),df.c1.max(),10)
bins
```

```python
# 구간 안에 속하느냐?
df['c1_bin'] = np.digitize(df['c1'], bins)
# 최댓값: 10, 최솟값: 1
```

```python
df
```

```python
df.groupby('c1_bin').size()
```

```python
df.groupby('c1_bin')['c1'].mean()
```

```python
df.groupby('c1_bin')['c2'].value_counts()
```

```python
# df의' c1_bin'열의 값이 2와 같은 데이터만 출력
df[df['c1_bin'] == 3]
```

```python
fruit = pd.DataFrame({'name':['apple','banana','cherry','durian',np.nan,'banana'],
             'color':['red','yellow','red','green',np.nan, 'yellow']})
```

```python
pd.get_dummies(fruit.columns=['name'], dummy_na=True)
```

```python
# np.where 연속형 변수 => 이산형화
# np.where(조건식, 'high','low')
# c1 열의 평균값보다 크면 'high', 작으면 'low' 값을 갖도록 값을 df['high_low'] 변수에 저장
df['high_low'] = np.where(df['c1']>df.c1.mean(),'high','low')
```

```python
# df['high_low'] 그룹화했을 때 각 그룹에 대한 크기?
df.groupby('high_low').size()
```

```python
q1 = np.percentile(df['c1'],25)
q3 = np.percentile(df['c1'],75)
```

```python
# df의 c1 값이 3사분위수보다 크면 'high', 1사분위수보다 크면'medium', 그렇지 않으면 'low'
df['hml'] = np.where((df['c1']>q3) ,'high', np.where(df['c1']>=q1, 'medium', 'low'))
```

```python
# 'hml' 컬럼값에 대해 그룹화, 각 그룹별 크기, 평균, 표준편차 출력
df.groupby('hml').size()
```

```python
df.groupby('hml').mean()
```

```python
df.groupby('hml').std()
```

```python
data = pd.DataFrame({'cust_id': ['c1', 'c1', 'c1', 'c2', 'c2', 'c2', 'c3', 'c3', 'c3'],
   'prod_cd': ['p1', 'p2', 'p3', 'p1', 'p2', 'p3', 'p1', 'p2', 'p3'],
   'grade' : ['A', 'A', 'A', 'A', 'A', 'A', 'B', 'B', 'B'],
   'pch_amt': [30, 10, 0, 40, 15, 30, 0, 0, 10]})
data
```

```python
# dataframe.pivot
df_pivot=data.pivot(index='cust_id',columns='prod_cd',values='pch_amt')
df_pivot
```

```python
# pivot에서는 에러가 나오지만 pivot_table에서는 에러가 나지 않는 경우가 있기때문에 그냥 pivot_table 써라.
df_pivot_t=pd.pivot_table(data, index='cust_id',columns='prod_cd',values='pch_amt')
df_pivot_t
```

```python
data
```

```python
# index: cust_id, grade
# columns: prod_cd
# values: pch_amt
df_pivot_t=pd.pivot_table(data, index=['cust_id', 'grade'],columns='prod_cd',values='pch_amt')
df_pivot_t
```

```python
df_pivot_table=pd.pivot_table(data, index=['cust_id'],columns=['grade','prod_cd'],values='pch_amt')
df_pivot_table
```

```python
data
```

```python
pd.pivot_table(data, index='grade', columns='prod_cd',values='pch_amt',aggfunc=np.sum, margins = True)
# 기본적으롷 aggfunc는 np.mean 적용되어있음., margins 적용하면 All 나옴
```

### melt: 재구조화

```python
data = pd.DataFrame({'cust_ID' : ['C_001', 'C_001', 'C_002', 'C_002'],
   'prd_CD' : ['P_001', 'P_002', 'P_001', 'P_002'],
   'pch_cnt' : [1, 2, 3, 4],
   'pch_amt' : [100, 200, 300, 400]})
data
```

```python
pd.melt(data, id_vars=['cust_ID','prd_CD'], )
```

```python
data_melt = pd.melt(data, id_vars=['cust_ID','prd_CD'], var_name='pch_CD', value_name='pch_value')
data_melt
```

```python
data_melt.index
data_melt.columns
```

```python
datamelt_pivot = pd.pivot_table(data_melt,index = ['cust_ID','prd_CD'],
              columns = 'pch_CD',
              values = 'pch_value')
datamelt_pivot
```
