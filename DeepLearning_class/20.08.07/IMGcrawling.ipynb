```python
from urllib.request import urlopen
from urllib.parse import quote_plus # 한글 >> 유니코드 자동 변환 
from bs4 import BeautifulSoup
baseUrl = 'https://search.naver.com/search.naver?where=image&sm=tab_jum&query='
word = input("검색어 입력: ") # ex) 강아지, 고양이 ...
num = int(input("크롤링 할 데이터 갯수: ")) # ex) 30

print(word)
print(quote_plus(word))

url = baseUrl + quote_plus(word)
html = urlopen(url)
print(html)

soup = BeautifulSoup(html, 'html.parser')
```

```python
img = soup.find_all(class_='_img')
img
```

```python
# 게시글의 다양한 정보 가져오기.
import re

def get_conten(driver):
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml') #html.parsar와 비슷한데 조금 더 가볍대? 근데 잘 모르겠어
    try:
        # 예외처리(try)
        content = soup.select('div.C4VMK > span')[0].text # 본문 내용 select
    except:   # 예기치 못한 상황이 발생하면 공백만 하나 출력할거야.
        content = ' '
    # content에서 해시태그된 내용 가져오기
    tags = re.findall("#[^\s]+",content) # \s: 공백 >> [^\s] 공백을 제외하고!! 대괄호 안에 ^는 반대
    
    # 작성일자 가져오기
    date = soup.select("time._1o9PC.Nzb55")[0]['datetime'][:10] # 리스트 1번째 요소의 datetime 속성의 값 추출
    
    #좋아요 수 가져와서 저장 ( like = ?)
    like = soup.select("button.sqdOP.yWX7d._8A5w5 > span")[0].string
    
    # 위치 정보( div.M30cS  ex. 어떤 장소가 많이 언급 됐냐? 등)
    place = soup.select("div.M30cS")[0].text
    
    data = [content, date, like, place, tags]
    return data
    
get_conten(driver)
```
