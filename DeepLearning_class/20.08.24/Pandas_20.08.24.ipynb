```python
time_zones = []  # 'tz'키에 대한 값을 추출하여 time_zone에 저장하는 코드 완성.
time_zones = [rec['tz'] for rec in records if 'tz' in rec]
time_zones
```

```python
len(time_zones)
```

```python
def get_counts(seq): # time_zones(3440개 대륙/도시)
    counts = {}
    for x in seq:
        if x in counts:# x('대륙/도시')가 counts에 key로 있다면
            counts[x] += 1
        else:
            counts[x] = 1  # 초기화
            
    return counts

get_counts(time_zones)
```

```python
def top_counts(count_dict, n = 10):
    vkpairs = [(count, tz) for tz, count in count_dict.items()]
    vkpairs.sort()
    print(vkpairs[-n:])
    
    # return # 상위 10개 도시

top_counts(get_counts(time_zones))
```
