### 2.1 날짜를 포멧에 맞추어 컬럼을 새로 추가

```python
def formatting(x):
    y = x[:4]
    m = x[4:]
    if m[0]=='0':
        res = y+'-'+m[1:]
    else:
        res=y+'-'+m
    return res
```

### 3.3 2019년 공휴일

```python
MON, TUE, WED, THU, FRI, SAT, SUN = range(7)
 
class holiday2020():
    HOLIDAYS = ((1,1), # new Year
                (1,24), # new Year0
                (1,25), # new Year1
                (1,26), # new Year2
                (3,1), # 3.1
                (4,15), # 21st Vote Day
                (4,30), # Buddha Day
                (5,5), # Children's Day
                (6,6), # Memorial Day
                (8,15), # Liberation Day
                (9,30), # Thanksgiving0
                (10,1), # Thanksgiving1
                (10,2), # Thanksgiving2
                (10,3), # National Foundation Day
                (10,9), # Hangul Day
                (12,25)) # Christmas
 
    def is_holiday(self, daytuple):
        HOLIDAYS = self.HOLIDAYS
        if daytuple in HOLIDAYS:
            return 1
        else:
            return 0
```

```python
l = []
if __name__=='__main__':
    for i in df_ccsi['방송일시']:
        now = i
        daytuple = ((now.month),(now.day))
        nowholiday = holiday2020()
        l.append(nowholiday.is_holiday(daytuple))
        
holiday = pd.Series(l).astype('category')
df_ccsi['휴일'] = holiday
```

- 24절기의 컬럼 0~345도를 일시 컬럼의 날짜에 대입 한뒤, 도별로 범위를 나눈 뒤 1,2,3... 으로 매핑

### 데이터 설명

get24DivisionInfo  
>월별로 구분(24절기), 요일, 공휴일 여부 등의 정보를 제공한다.  

- https://m.blog.naver.com/k5808151/221625545385 (절기별 태양황경)  
- http://www.linuxchannel.net/docs/solar-24terms.txt (참고)

```python
driver=webdriver.Chrome("C:/mychrome/chromedriver")
```

```python
service_key="공공데이터포털키"
```

```python
data=[]
for num in range(1,13):
    month="{:02d}".format(num)
    url="http://apis.data.go.kr/B090041/openapi/service/SpcdeInfoService/get24DivisionsInfo?solYear=2019&solMonth={}&ServiceKey={}".format(month,service_key)
    driver.get(url)
    html=driver.page_source
    soup=BeautifulSoup(html, "xml")
    time.sleep(2)
    for i in range(len(soup.select("item"))):
        dateName=soup.select("item > dateName")[i].text.strip()
        locdate=soup.select("item > locdate")[i].text.strip()
        sunLongitude=soup.select("item > sunLongitude")[i].text.strip()
        print(dateName, locdate, sunLongitude)
        data.append([locdate, dateName, sunLongitude])
        time.sleep(1)
```

```python
solarTerm=pd.DataFrame(data, columns=['날짜','명칭','태양황경'])
solarTerm.to_csv("solarTerm", index=False)
solarTerm
```

```python
# prime 시간대:
# 평일: 07:00~11:00, 15:00~18:00
# 주말: 09:00~12:00, 16:00~19:00, 21:00
checking['prime'] = 0

for i in list(checking.index):
    # 평일
    if checking.loc[i, 'resting_day']==False:
        if (checking.loc[i, 'hour'] >= 7) & (checking.loc[i, 'hour'] <= 11):
            checking.loc[i, 'prime'] = 1
        elif (checking.loc[i, 'hour'] >= 15) & (checking.loc[i, 'hour'] <= 18):
            checking.loc[i, 'prime'] = 1
    # 주말
    if checking.loc[i, 'resting_day']==True:
        if (checking.loc[i, 'hour'] >= 9) & (checking.loc[i, 'hour'] <= 12):
            checking.loc[i, 'prime'] = 1
        elif (checking.loc[i, 'hour'] >= 16) & (checking.loc[i, 'hour'] <= 19):
            checking.loc[i, 'prime'] = 1
        elif checking.loc[i, 'hour'] == 21:
            checking.loc[i, 'prime'] = 1
```

```python
rfr=RandomForestRegressor(n_estimators=200, random_state=42)
```

```python
rfr.fit(xTrain, yTrain)
```

```python
yPred=rfr.predict(xTest)
```

```python
def percentage_error(actual, predicted):
    res = np.empty(actual.shape)
    for j in range(actual.shape[0]):
        if actual[j] != 0:
            res[j] = (actual[j] - predicted[j]) / actual[j]
        else:
            res[j] = predicted[j] / np.mean(actual)
    print(res)
    return res

def mean_absolute_percentage_error(y_true, y_pred): 
    return np.mean(np.abs(percentage_error(np.asarray(y_true), np.asarray(y_pred)))) * 100
```

```python
rfr.score(xTrain,yTrain)
```

```python
mean_absolute_percentage_error(yTest,yPred)
```
