```python
from selenium import webdriver
from bs4 import BeautifulSoup
import re
import time
```

```python
driver=webdriver.Chrome("/Users/jinwon-kim/Developer/chromedriver")
```

```python
service_key="api서비스키"
```

```python
data=[]
for num in range(1,13):
    month="{:02d}".format(num)
    url="http://apis.data.go.kr/B090041/openapi/service/SpcdeInfoService/get24DivisionsInfo?solYear=2020&solMonth={}&ServiceKey={}".format(month,service_key)
    driver.get(url)
    html=driver.page_source
    soup=BeautifulSoup(html, "xml")
    time.sleep(2)
    for i in range(len(soup.select("item"))):
        dateName=soup.select("item > dateName")[i].text.strip()
        locdate=soup.select("item > locdate")[i].text.strip()
        sunLongitude=soup.select("item > sunLongitude")[i].text.strip()
        print(dateName, locdate, sunLongitude)
        data.append([locdate, dateName, sunLongitude])
        time.sleep(1)
```

```python
solarTerm=pd.DataFrame(data, columns=['날짜','명칭','태양황경'])
solarTerm.to_csv("solarTerm2020", index=False)
```

```python
solarTerm_df=pd.read_csv("solarTerm2020")
solarTerm_df['날짜']=pd.to_datetime(solarTerm_df['날짜'],format='%Y%m%d')
```
